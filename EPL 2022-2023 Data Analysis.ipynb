```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
player = pd.read_csv('all stats premier league 22-23.csv')
```

```python
player.drop(['Unnamed: 0', 'Matches'], axis=1)
```

```python
player.columns
```

```python
#drop column
player = player[['Player', 'Nation', 'Pos', 'Age',
        'Gls', 'Ast', 'G+A', 'G-PK', 'PK', 'PKatt', 'CrdY', 'CrdR', 'xG',
       'npxG', 'xAG', 'npxG+xAG', 'PrgC', 'PrgP', 'PrgR', 'Club',
       'Tkl', 'TklW', 'Def 3rd', 'Mid 3rd', 'Att 3rd', 'Tkl.1', 'Att', 'Tkl%',
       'Lost', 'Blocks', 'Sh', 'Pass', 'Int', 'Tkl+Int', 'Clr', 'Err']]
player.head()
```

```python
#cleaning club data
player['Club'] = [c.replace('Stats', '').replace('-', ' ') for c in player['Club']]
player.head()
```

## Top 5 player nationality

```python
#player by nationality top 5
most_nation = player.groupby('Nation').size().sort_values(ascending=False)
most_nation.head(5).plot(kind='bar')
```

## Club with most goal

```python
#top 5 goal by club 
goal_by_club = player.groupby('Club')['Gls'].sum().sort_values(ascending=False).reset_index()
plt.figure(figsize=(15,8))
sns.barplot(data=goal_by_club.head(5), x='Club', y='Gls')
```

## Top 5 Goal Scorer

```python
#goal by player
player_goal = player.groupby('Player')['Gls'].sum().sort_values(ascending=False).reset_index()
plt.figure(figsize=(10,8))
sns.barplot(data=player_goal.head(5), x='Player', y='Gls')
```

## Most assist by player

```python
#Player with most assist
player_assist = player.groupby('Player')['Ast'].sum().sort_values(ascending=False).reset_index()
plt.figure(figsize=(10,8))
sns.barplot(data=player_assist.head(5), x='Player', y='Ast')
```

## Teams with most xG and Goals

```python
xg_goals = player.groupby('Club')[['Gls', 'xG']].sum().sort_values(ascending=False, by='Gls').reset_index()
sns.scatterplot(data=xg_goals, x='xG', y='Gls', hue='Club')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
```

## Club with most Red Cards

```python
most_red_card = player.groupby('Club')['CrdR'].sum().sort_values(ascending=False).reset_index()
plt.figure(figsize=(12,8))
sns.barplot(data=most_red_card.head(5), x='Club', y='CrdR')
```

## Most assist and xA

```python
xa_assist = player.groupby('Club')[['Ast', 'xAG']].sum().sort_values(ascending=False, by='Ast').reset_index()
sns.scatterplot(data=xa_assist, x='xAG', y='Ast', hue='Club')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
```

## Penalty vs Pen Attempt

```python
penalty_att = player.groupby('Club')['PK', 'PKatt'].sum().sort_values(ascending=False, by='PK').reset_index()
sns.scatterplot(data=penalty_att, x='PK', y='PKatt', hue='Club')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
```
